```python
from urllib.request import urlopen
```

```python
url = 'https://www.flipkart.com/'
```

```python
resp = urlopen(url)
```

```python
resp.status
```

```python
Content = resp.read()
```

```python
Content
```

```python
dcontent = Content.decode('utf-8')
```

```python
print(dcontent)
```
